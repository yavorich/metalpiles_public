```python
import os
import glob
import numpy as np
import base64
import json
import cv2
from matplotlib import pyplot as plt
```

```python
FRAMES_PATH = os.path.join('..', 'data', 'frames(1)')
```

```python
json_list = glob.glob(os.path.join(FRAMES_PATH, '*.json'))
json_list[0:5]
```

```python
def b64_to_img(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)
        img = data['imageData']
        img = base64.b64decode(img)
        img = np.frombuffer(img, dtype = np.uint8)
        img = cv2.imdecode(img, 1)
        return img
```

Пример старой картинки

```python
fig = plt.figure(figsize=(16*2, 9*2))
plt.imshow(b64_to_img(json_list[200]))
plt.show()
```

Переопределение imageData в json файлах

```python
frames_type = '.jpg'

for json_path in json_list:
    with open(json_path, 'r') as json_f:
#         print(f'Reassigning {json_path.split(os.path.sep)[-1]}')
#       loading json file
        json_data = json.load(json_f)
        
#       Encoding image and reassigning a new img value
        jpg_path = json_path[:-5] + frames_type
        img = cv2.imread(jpg_path)
        retval, buffer = cv2.imencode(frames_type, img)
        b64_img = base64.b64encode(buffer)
        b64_img = b64_img.decode('utf-8')
        old_img_data = json_data['imageData']
        json_data['imageData'] = b64_img

#       Save results
    with open(json_path, 'w') as write_f:    
        json.dump(json_data, write_f)
    
    print(f'{json_path.split(os.path.sep)[-1]} reassigned')
```
